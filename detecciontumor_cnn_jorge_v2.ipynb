```python
def imshow(img):
    fig,ax = plt.subplots(1,1,figsize=(8,8))
    ax.imshow(img,cmap='gray')
```

```python
possible_labels=os.listdir(paths[0])
```

```python
possible_labels
```

```python
paths = [path if path.endswith('/') else path + '/' for path in paths]
```

```python
data = []

for path in paths:
    for label_int, label_string in enumerate(possible_labels):
        full_path = path + label_string + '/'
        print(f"Buscando en: {full_path}")
        for filename in glob.glob(full_path + '*.jpg'):
            print(f"Leyendo archivo: {filename}")
            img = cv2.imread(filename)
            if img is not None:
                data.append([label_int, cv2.resize(img, (320, 320))])
            else:
                print(f"Error leyendo la imagen: {filename}")

print(f"Total de datos cargados: {len(data)}")
```

```python
import random
random.Random(0).shuffle(data)

x_train=[]
y_train=[]

x_val=[]
y_val=[]

x_test=[]
y_test=[]

for i, sample in enumerate(data):
    label=sample[0]
    img=sample[1]
    if i<= 0.8*len(data):
        x_train.append(img)
        y_train.append(label)
    elif i>0.8*len(data) and i<=0.9*len(data):
        x_val.append(img)
        y_val.append(label)
    else:
        x_test.append(img)
        y_test.append(label)

x_train=np.array(x_train)
x_val=np.array(x_val)
x_test=np.array(x_test)

y_train=np.array(y_train)
y_val=np.array(y_val)
y_test=np.array(y_test)
```

# **2. Definici√≥n y entrenamiento de la red neuronal convolucional**

```python
import tensorflow as tf
y_trainOneHot=tf.one_hot(y_train,len(possible_labels))
y_valOneHot=tf.one_hot(y_val,len(possible_labels))
y_testOneHot=tf.one_hot(y_test,len(possible_labels))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
def cnn_model():
    model = Sequential()
    model.add(Convolution2D(32,(3,3),input_shape=(320,320,3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Flatten())
    model.add(Dense(30,activation='relu'))
    model.add(Dense(4,activation='softmax'))
    model.compile(optimizer='SGD', loss='categorical_crossentropy',metrics=['accuracy'])
    return model
```

```python
model=cnn_model()
model.summary()
```

```python
from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint

# Definir el callback para guardar el mejor modelo basado en la m√©trica deseada
checkpoint = ModelCheckpoint(filepath='best_model.keras', 
                             monitor='val_accuracy',  # Puedes cambiar 'val_accuracy' por otra m√©trica si lo deseas
                             mode='max',  # 'max' si la m√©trica deseada debe ser maximizada, 'min' si debe ser minimizada
                             save_best_only=True)
```
